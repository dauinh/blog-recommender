```python
import pandas as pd
```

```python
blog_df = pd.read_csv('Medium Blog Data.csv')
author_df = pd.read_csv('Author Data.csv')
ratings_df = pd.read_csv('Blog Ratings.csv')
```

```python
blog_df = blog_df.drop(columns=['author_id', 'scrape_time'])
blog_df
```

```python
five_stars_df = ratings_df[ratings_df['ratings'] == 5.0]
user_group = five_stars_df.groupby('userId')
user_group.head()
```

```python
user_history = user_group['blog_id'].apply(list)
user_history
```

```python
user_topic = {}
for user_id, group in user_group:
    merged_group = pd.merge(group, blog_df, on='blog_id')
    user_topic[user_id] = merged_group['topic'].tolist()
    user_topic[user_id] = list(set(user_topic[user_id]))

user_topic = pd.Series(user_topic)
user_topic
```

```python
user_df = pd.DataFrame({
    'user_id': user_group.grouper.group_keys_seq,
    'topics': user_topic,
    'history': user_history
})
user_df
```

```python
user_df.to_csv('users.csv')
blog_df.to_csv('blogs.csv')
```
